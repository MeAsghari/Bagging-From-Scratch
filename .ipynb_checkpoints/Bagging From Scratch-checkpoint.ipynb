```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression , LinearRegression , SGDRegressor
from sklearn.neighbors import KNeighborsClassifier , KNeighborsRegressor
from sklearn.metrics import accuracy_score , r2_score
from mlxtend.plotting import plot_decision_regions
from sklearn.tree import DecisionTreeClassifier , DecisionTreeRegressor
from sklearn import svm

from scipy.stats import mode
```

## **Classification Bagging**

```python
df = pd.read_csv('data/2class_circle_data.csv' , header=None)
df = df.T
df
```

```python
X = df.iloc[: , :2].values
y = df.iloc[: , 2].values
X.shape , y.shape
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y ,test_size=0.1 , random_state=0 , stratify=y)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
fig , ax = plt.subplots(1 , 2 , figsize=(14 , 5))
axes = ax.flatten()


axes[0].scatter(X_train[: , 0] , X_train[: , 1] , c=y_train , s=60 , cmap='jet', ec='k')
axes[0].grid()

axes[1].scatter(X_test[: , 0] , X_test[: , 1] , c=y_test , s=60 , cmap='jet' , ec='k');
axes[1].grid()
```

```python
y_preds = []
models = []
for i in range(21):
    
    model = KNeighborsClassifier(n_neighbors=5)
    X_subset , _ , y_subset , _ = train_test_split(X_train , y_train , test_size=0.6 , stratify=y_train)
    model.fit(X_subset , y_subset)
    models.append(model)
    y_pred = model.predict(X_train)
    y_pred = list(y_pred)
    y_preds.append(y_pred)
    print(f'Score for model {i} is: {accuracy_score(y_train , y_pred)}')

ensemble_prediction = mode(y_preds).mode.ravel()
print(f'Score for ensemble model is: {accuracy_score(y_train , ensemble_prediction)}')
```

```python
x1 = np.linspace(X_train[: ,0].min() , X_train[:, 0].max() , 200)
x2 = np.linspace(X_train[: ,1].min() , X_train[:, 1].max() , 200)

x1m , x2m = np.meshgrid(x1 , x2)
```

```python
x1r = x1m.ravel()
x2r = x2m.ravel()

Xm = np.stack((x1r , x2r))
Xm = Xm.T
```

```python
fig , axes = plt.subplots(7 , 3 , figsize=(30 , 25))
axes_flat = axes.flatten()

all_preds = []
for i , ax in enumerate(axes_flat):
    
    if i == 21:
        break
        
    prediction = models[i].predict(Xm)
    
    sns.scatterplot(x=Xm[: , 0] , y=Xm[: , 1] , c=prediction , alpha=0.1 , s=60 , cmap='jet' , ax=ax);
    sns.scatterplot(x=X_train[: , 0] , y=X_train[: , 1] , c=y_train , s=60 , cmap='jet', ec='w' , ax=ax);
    
    prediction = list(prediction)
    all_preds.append(prediction)
```

```python
ensemble_prediction = mode(all_preds).mode.ravel()
sns.scatterplot(x=Xm[: , 0] , y=Xm[: , 1] , c=ensemble_prediction , alpha=0.1 , s=60 , cmap='jet');
sns.scatterplot(x=X_train[: , 0] , y=X_train[: , 1] , c=y_train , s=60 , cmap='jet', ec='w');
```

## **Regression Bagging**

```python
train_set = pd.read_csv('data/energy-train-l.csv')
test_set = pd.read_csv('data/energy-test-l.csv')
train_set.shape , test_set.shape
```

```python
plt.figure(figsize=(7 , 5))
plt.scatter(train_set['Temperature'] , train_set['Demand'] , c='b' , ec='k' , label='train set');
plt.scatter(test_set['Temperature'] , test_set['Demand'] , c='r' , ec='k' , label = 'test set');
plt.legend();
```

```python
X_train = train_set.iloc[: , 0].values
y_train = train_set.iloc[: , -1].values

X_test = test_set.iloc[: , 0].values
y_test = test_set.iloc[: , -1].values

X_train = X_train[: , None]
X_test = X_test[: , None]

print(X_train.shape)
print(y_train.shape)

print(X_test.shape)
print(y_test.shape)
```

```python
y_preds = []
models = []
for i in range(21):
    
    #model = KNeighborsRegressor(n_neighbors=4)
    model = DecisionTreeRegressor(max_depth=3)
    X_subset , _ , y_subset , _ = train_test_split(X_train , y_train , test_size=0.6)
    model.fit(X_subset , y_subset)
    models.append(model)
    y_pred = model.predict(X_train)
    y_pred = list(y_pred)
    y_preds.append(y_pred)
    print(f'Score for model {i} is: {r2_score(y_train , y_pred)}')

    
ensemble_prediction = np.mean(y_preds , axis=0)
print(f'Score for ensemble model is: {r2_score(y_train , ensemble_prediction)}')
```

```python
xp = np.linspace(X_train.min(), X_train.max(), 1000)[:, None]

fig , axes = plt.subplots(7 , 3 , figsize=(25 , 30))
axes_flat = axes.flatten()

all_preds = []
for i , ax in enumerate(axes_flat):
    
    if i == 21:
        break
        
    prediction = models[i].predict(xp)
    
    ax.scatter(train_set['Temperature'] , train_set['Demand'] , c='b' , ec='k' , label='train set');
    ax.scatter(test_set['Temperature'] , test_set['Demand'] , c='r' , ec='k' , label = 'test set');
    ax.plot(xp, prediction, c='gold', linewidth=3 , label='Predict Line');
    ax.legend()
    
    prediction = list(prediction)
    all_preds.append(prediction)
```

```python
ensemble_prediction = np.mean(all_preds , axis=0)
plt.scatter(train_set['Temperature'] , train_set['Demand'] , c='b' , ec='k' , label='train set');
plt.scatter(test_set['Temperature'] , test_set['Demand'] , c='r' , ec='k' , label = 'test set');
plt.plot(xp, ensemble_prediction, c='gold', linewidth=3 , label='Predict Line');
```
